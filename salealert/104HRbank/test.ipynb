```python
def pretreat(df):
        def work_content(s):
            for e in ["保險業務員","壽險顧問","保險儲備主管","理財"]:
                if e in s:
                    return False
                else:
                    return True

        def work_time(s):
            pattern="\d+:\d+"
            r=re.findall(pattern, s, flags=0)
            if len(r) == 2:
                if (int(r[1].replace(":",""))-int(r[0].replace(":","")))<800:
                    return False
                else:
                    return True
            else:
                return True

        def work_salary(s):
            return (s == '時薪158元' or s == '待遇面議')

        def work_com_num(s):
            if '核准文號' in s:
                patern="[\dA-Z]+"
                r=re.findall(patern,s)
                if r:
                    if r[0] not in ["109A01","109A02","109A03"]:
                        return False
                    else:
                        return True
                else:
                    return True
            else :
                return True
        df["是否違規"]=~(df["工作名稱"].apply(work_content)&df["上班時段"].apply(work_time)&df['工作待遇'].apply(work_salary)&df["公司"].apply(work_com_num))
        df['違規原因']=np.nan
        df.loc[~df["工作名稱"].apply(work_content),'違規原因']="工作名稱"
        df.loc[~df["上班時段"].apply(work_time),'違規原因']="上班時段錯誤(未滿8小時的時間)"
        df.loc[~df['工作待遇'].apply(work_salary),'違規原因']="薪資福利錯誤"
        df.loc[~df["公司"].apply(work_com_num),'違規原因']="核准文號錯誤"
        return df
```

```python
import numpy as np
df = pretreat(df)
```
